#1.
Case Sensitivity: If your preprocessing engine does not include a "lowercasing" step, the model treats "Love" and "love" as two completely different features in its vocabulary matrix.

#2
Dimensionality Explosion: Your feature set becomes bloated. Since these words appear in almost every document, they take up significant space in your word-frequency matrices without helping distinguish one document from another.

#3
 Consider the phrase "The movie was not good." If you remove "not" (a common stopword), the engine sees "movie good" and predicts a positive sentiment. You’ve effectively flipped the meaning of the sentence.

#4
Both techniques aim to reduce a word to its base form, but they use very different "surgical" methods.

Stemming (The Machete): It follows a set of heuristic rules to chop off the ends of words. It’s fast but "dumb."

Example: "Caring" and "Cars" might both be chopped to "Car." It often results in non-words (e.g., "Studies" becomes "studi").

Lemmatization (The Scalpel): It uses a dictionary (morphological analysis) and considers the word's Part-of-Speech (POS).

Example: It knows that "saw" could be the past tense of "see" (verb) or a tool for wood (noun). It returns a valid dictionary word (the "lemma").

**Task 2:Build Advanced Preprocessing Function**

```python
import re
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\S+@\S+\.\S+', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'\d+', '', text)
    raw_tokens = re.findall(r'\b\w+\b', text)
    meaningful_short_words = {'no', 'not'}
    
    filtered_tokens = [
        token for token in raw_tokens 
        if len(token) > 2 or token in meaningful_short_words
    ]
    return " ".join(filtered_tokens)
examples = [
    "I have 2 dogs",
    "This   is    good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com or email me at test@dev.com now",
    "I do not like it, no."
]

for ex in examples:
    print(f"Original: {ex}")
    print(f"Processed: {preprocess_text(ex)}\n")
```

**Task 6: Build Full Pipeline**

```python
import re

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    meaningful_short_words = {'no', 'not'}

    for text in text_list:
        processed = text.lower()
        processed = re.sub(r'https?://\S+|www\.\S+|\S+@\S+\.\S+', '', processed)
        processed = re.sub(r'(.)\1{2,}', r'\1\1', processed)
        processed = re.sub(r'\d+', '', processed)
        raw_tokens = re.findall(r'\b\w+\b', processed)
        current_tokens = [
            token for token in raw_tokens 
            if len(token) > 2 or token in meaningful_short_words
        ]
        all_tokens.extend(current_tokens)
        clean_sentences.append(" ".join(current_tokens))
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

raw_data = [
    "I have 2 dogs and 1 cat!!",
    "Visit http://google.com for info",
    "Nooooo, I am not happy with the service.",
    "This is sooooo goooood!!!"
]

output = full_pipeline(raw_data)

# Displaying the structured output
import json
print(json.dumps(output, indent=4))
```

**Task 7: Error Handling**

```python
import re

def full_pipeline(text_list):
    if not isinstance(text_list, list):
        return {"tokens": [], "clean_sentences": []}

    all_tokens = []
    clean_sentences = []
    meaningful_short_words = {'no', 'not'}

    for text in text_list:
        if not isinstance(text, str) or not text.strip():
            clean_sentences.append("")
            continue
        processed = text.lower()
        processed = re.sub(r'https?://\S+|www\.\S+|\S+@\S+\.\S+', '', processed)
        processed = re.sub(r'(.)\1{2,}', r'\1\1', processed)
        processed = re.sub(r'\d+', '', processed)
        raw_tokens = re.findall(r'\b\w+\b', processed)
        current_tokens = [
            token for token in raw_tokens 
            if len(token) > 2 or token in meaningful_short_words
        ]
        if not current_tokens:
            clean_sentences.append("")
        else:
            all_tokens.extend(current_tokens)
            clean_sentences.append(" ".join(current_tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
my_test_data = [
    "I have 2 dogs and 1 cat!!",
    "Visit http://google.com for info",
    "Nooooo, I am not happy with the service.",
    "This is sooooo goooood!!!",
    "   ",          
    "999888",       
    "😍😍😍"         
]
results = full_pipeline(my_test_data)
import json
print(json.dumps(results, indent=4))
```
